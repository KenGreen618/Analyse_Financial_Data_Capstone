```python
# calculating daily mean returns
daily_mean_returns = etf_returns_df.mean()
print(daily_mean_returns)
daily_mean_returns = daily_mean_returns.sort_values()
print(daily_mean_returns)
```

```python
# daily mean index for x axis
daily_mean_returns.keys()

# grab each daily mean value for the y axis
height = []
for key in daily_mean_returns.keys():
    height.append(daily_mean_returns[key])

# arrange keys on x acis based on length
x_pos = np.arange(len(daily_mean_returns.keys()))

# plot bar chart
plt.bar(x_pos, height)
plt.xticks(x_pos, daily_mean_returns.keys())
plt.xlabel("Sector ETFs")
plt.ylabel("Daily mean")
plt.title("Daily mean rate of return")

plt.show()
```

```python
# daily mean index for x axis
daily_mean_returns.keys()

# grab each daily mean value for the y axis
height = []
for key in daily_mean_returns.keys():
    height.append(daily_mean_returns[key]*252)

# arrange keys on x acis based on length
x_pos = np.arange(len(daily_mean_returns.keys()))

# plot bar chart
plt.bar(x_pos, height)
plt.xticks(x_pos, daily_mean_returns.keys())
plt.xlabel("Sector ETFs")
plt.ylabel("Daily mean")
plt.title("Annualised mean rate of return")

plt.show()
```

```python
# calcuate variance 
daily_variance = etf_returns_df.var()
print(daily_variance)
daily_variance = daily_variance.sort_values()
print(daily_variance)

# variance index for the x axis 
daily_variance.keys()

# grab each variance value for the y axis
height = []
for key in daily_variance.keys():
    height.append(daily_variance[key])

# arrange keys on x axis based on length
x_pos = np.arange(len(daily_variance.keys()))

# plot bar chart
plt.bar(x_pos, height)
plt.title('Sector ETFs variance comparison')
plt.xticks(x_pos, daily_variance.keys())
plt.xlabel("Sector ETFs")
plt.ylabel("Daily Variance")
plt.show()
```

```python
# calculate standard deviation
daily_std = etf_returns_df.std()
daily_std = daily_std.sort_values()
print(daily_std)
daily_std.keys()
height = []
for key in daily_std.keys():
    height.append(daily_std[key])
x_pos = np.arange(len(daily_std.keys()))
plt.bar(x_pos, height)
plt.xticks(x_pos, daily_std.keys())
plt.xlabel("Sector ETFs")
plt.ylabel("std")
plt.title("Daily std")
plt.show()
```

```python
etf_correlations = etf_returns_df.corr().abs()
print(etf_correlations)
s = etf_correlations.unstack()
so = s.sort_values()
print('sorted lowest correlations')
print(so)

def get_redundant_pairs(input_df):
    #get diagonal and lower triangular pairs of correlation matrix
    pairs_to_drop = set()
    cols = input_df.columns
    for i in range(0, input_df.shape[1]):
        for j in range(0, i+1):
            pairs_to_drop.add((cols[i], cols[j]))
    return pairs_to_drop

def get_top_abs_correlations(input_df, n=8):
    au_corr = input_df.corr().abs().unstack()
    labels_to_drop = get_redundant_pairs(input_df)
    au_corr = au_corr.drop(labels=labels_to_drop).sort_values(ascending=False)
    return au_corr[0:n]

print('top 5 highest correlations')
print(get_top_abs_correlations(etf_returns_df, 5))
        
```
